### Analysis

- Most effective treatment regimens are Capomulin and Ramicane, with no outliers.
- Overall Capomulin appears to be the less effective of the two.

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np
from scipy.stats import linregress
import seaborn as sns

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single DataFrame
combined_data = pd.merge(study_results,mouse_metadata, on='Mouse ID')


# Display the data table for preview
combined_data
```

```python
# Checking the number of mice.
mouse_count = combined_data['Mouse ID'].nunique()
mouse_count
```

```python
# Find duplicate mice based on Mouse ID and Timepoint
duplicate_mice = combined_data[combined_data.duplicated(subset=['Mouse ID', 'Timepoint'], keep=False)]

# Get the unique Mouse IDs that are duplicated
duplicate_mouse_ids = duplicate_mice['Mouse ID'].unique()
duplicate_mouse_ids
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
# Identify the duplicate mouse ID
duplicate_mouse_id = duplicate_mice['Mouse ID'].unique()[0]

# Get all data for the duplicate mouse ID
duplicate_mouse_data = combined_data[combined_data['Mouse ID'] == duplicate_mouse_id]

# Display the data for the duplicate mouse ID
duplicate_mouse_data
```

```python
# Create a clean DataFrame by dropping the duplicate mice by their IDs
# Using ~ to negate boolean series so if Mouse ID is in the duplicate_mouse_ids DF it's removed from the cleaned dataframe
cleaned_data = combined_data[~combined_data['Mouse ID'].isin(duplicate_mouse_ids)]
cleaned_data.head()
```

```python
# Checking the number of mice in the clean DataFrame.
mouse_count2 = cleaned_data['Mouse ID'].nunique()
mouse_count2
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
# Assemble the resulting series into a single summary DataFrame.
summary_stats = cleaned_data.groupby('Drug Regimen')['Tumor Volume (mm3)'].agg(['mean', 'median', 'var', 'std', 'sem'])

# Rename columns for clarity
summary_stats = summary_stats.rename(columns={
    'mean': 'Mean Tumor Volume',
    'median': 'Median Tumor Volume',
    'var': 'Tumor Volume Variance',
    'std': 'Tumor Volume Std. Dev.',
    'sem': 'Tumor Volume SEM'
})

# Display the summary statistics table
summary_stats
```

```python
# A more advanced method to generate a summary statistics table of mean, median, variance, standard deviation,
# and SEM of the tumor volume for each regimen (only one method is required in the solution)

# Using the aggregation method, produce the same summary statistics in a single line
summary_stats_advanced = cleaned_data.groupby('Drug Regimen')['Tumor Volume (mm3)'].agg(['mean', 'median', 'var', 'std', 'sem'])\
    .rename(columns={'mean': 'Mean (mm3)', 
                     'median': 'Median (mm3)',
                     'var': 'Variance (mm3)',
                     'std': 'Std. Dev. (mm3)', 
                     'sem': 'SEM (mm3)'})

# Display the advanced summary statistics table
summary_stats_advanced

```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin

selected_regimens = ['Capomulin', 'Ramicane', 'Infubinol', 'Ceftamin']
```

```python
# Filter the data for the selected regimens
final_tumor_data = cleaned_data[cleaned_data['Drug Regimen'].isin(selected_regimens)]


# Start by getting the last (greatest) timepoint for each mouse
last_timepoint = final_tumor_data.groupby('Mouse ID')['Timepoint'].max()
last_timepoint_df = pd.DataFrame(last_timepoint).reset_index()

# Merge this group df with the original DataFrame to get the tumor volume at the last timepoint
final_tumor_volume = pd.merge(last_timepoint_df, cleaned_data, on=['Mouse ID', 'Timepoint'], how='left')

# Display the final tumor volume data
final_tumor_volume.head()
```

```python
# Create empty list to fill with tumor vol data (for plotting)
tumor_vol_data = []

# Iterate through each regimen
for regimen in selected_regimens:
    # Locate the rows which contain mice on each drug and get the tumor volumes
    tumor_vol_subset = final_tumor_volume.loc[final_tumor_volume['Drug Regimen'] == regimen, 'Tumor Volume (mm3)']
    
    # Add subset to the list
    tumor_vol_data.append(tumor_vol_subset)
    
# Calculate the IQR and quantitatively determine if there are any potential outliers. 
# Locate the rows which contain mice on each drug and get the tumor volumes
# add subset 
    quartiles = tumor_vol_subset.quantile([0.25, 0.5, 0.75])
    lower_q = quartiles[0.25]
    upper_q = quartiles[0.75]
    iqr = upper_q - lower_q

    
    # Determine outliers using upper and lower bounds
    lower_bound = lower_q - 1.5 * iqr
    upper_bound = upper_q + 1.5 * iqr
    
    # Identify potential outliers
    outliers = tumor_vol_subset[(tumor_vol_subset < lower_bound) | (tumor_vol_subset > upper_bound)]
    
    # Print results
    print(f"{regimen} potential outliers: {outliers.tolist()}")

# Display the tumor volume data for each regimen
```

```python
# Generate a box plot that shows the distrubution of the tumor volume for each treatment group.
# Create a list to store tumor volume data for plotting
tumor_vol_data = []

# Iterate through each regimen
for regimen in selected_regimens:
    # Locate the rows which contain mice on each drug and get the tumor volumes
    tumor_vol_subset = final_tumor_volume.loc[final_tumor_volume['Drug Regimen'] == regimen, 'Tumor Volume (mm3)']
    # Add subset to the list
    tumor_vol_data.append(tumor_vol_subset)

# Create a box plot
plt.figure(figsize=(7,5))
plt.boxplot(tumor_vol_data, labels=selected_regimens, sym='ro')

# Add labels and title
plt.title('Distribution of Tumor Volume for Each Treatment Group')
plt.xlabel('Drug Regimen')
plt.ylabel('Tumor Volume (mm3)')

# Display the plot
plt.show()
```

```python
# Generate a scatter plot of mouse weight vs. the average observed tumor volume for the entire Capomulin regimen
# Filter data for Capomulin regimen
capomulin_data = cleaned_data[cleaned_data['Drug Regimen'] == 'Capomulin']

# Calculate the average observed tumor volume for each mouse
average_tumor_volume = capomulin_data.groupby('Mouse ID')['Tumor Volume (mm3)'].mean()

# Merge average tumor volume data with the mouse weight data
merged_data = pd.merge(average_tumor_volume, mouse_metadata, on='Mouse ID', how='inner')

# Create a scatter plot of mouse weight vs. average observed tumor volume
plt.figure(figsize=(7, 5))
plt.scatter(merged_data['Weight (g)'], merged_data['Tumor Volume (mm3)'], color='gold', alpha=0.7, edgecolors='black')

# Add labels and title
plt.title('Mouse Weight vs. Average Tumor Volume (Capomulin)')
plt.xlabel('Mouse Weight (g)')
plt.ylabel('Average Tumor Volume (mm3)')

# Display the plot
plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and a linear regression model 
# for mouse weight and average observed tumor volume for the entire Capomulin regimen


# Calculate the average observed tumor volume for each mouse
average_tumor_volume = capomulin_data.groupby('Mouse ID')['Tumor Volume (mm3)'].mean()

# Merge average tumor volume data with the mouse weight data
merged_data = pd.merge(average_tumor_volume, mouse_metadata, on='Mouse ID', how='inner')

# Calculate the correlation coefficient
correlation_coefficient = merged_data['Weight (g)'].corr(merged_data['Tumor Volume (mm3)'])

# Perform linear regression
slope, intercept, r_value, p_value, std_err = linregress(merged_data['Weight (g)'], merged_data['Tumor Volume (mm3)'])

# Create a scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(merged_data['Weight (g)'], merged_data['Tumor Volume (mm3)'], color='steelblue', alpha=0.9)

# Plot the linear regression line
regress_values = merged_data['Weight (g)'] * slope + intercept
plt.plot(merged_data['Weight (g)'], regress_values, color='r', linestyle='--', linewidth=2)

# Add labels and title
plt.title('Mouse Weight vs. Average Tumor Volume (Capomulin Regimen)')
plt.xlabel('Mouse Weight (g)')
plt.ylabel('Average Tumor Volume (mm3)')

# Display the plot
plt.show()
```

```python
# Display correlation coefficient and regression information
print(f"Correlation Coefficient: {correlation_coefficient}")
print(f"Linear Regression Model: y = {slope:.2f}x + {intercept:.2f}")
```
